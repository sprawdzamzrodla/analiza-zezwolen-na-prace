```python
import pandas as pd
```

```python
def read_data_file_1(year: int, tab: str, start: int) -> pd.DataFrame:
    _df = pd.read_excel(f"./data/Dane zbiorcze {year} r.xls", tab).iloc[start:,:2]
    _df.columns = ["kraj", "liczba zezwoleń"]
    _df = _df[_df['kraj'].str.upper() != _df['kraj']]
    _df["rok"] = year
    return _df.reset_index(drop=True)


def _preprocess_data_file_2(_df: pd.DataFrame, year: int) -> pd.DataFrame:
    _df = _df.fillna(method='ffill')
    _df.columns = ["kraj", "typ", "liczba zezwoleń"]
    _df['typ'] = _df['typ'].str.strip()
    _df = _df[_df['typ'] == "Razem"]
    _df = _df[_df['kraj'].str.upper() != _df['kraj']]
    _df = _df[["kraj", "liczba zezwoleń"]]
    _df["rok"] = year
    return _df.reset_index(drop=True)


def read_data_file_2(year: int, tab: str, start_row: int, start_column:int=0, file_format:str="xls") -> pd.DataFrame:
    _df = pd.read_excel(f"./data/Dane zbiorcze {year} r.{file_format}", tab).iloc[start_row:,start_column:start_column+3]
    return _preprocess_data_file_2(_df, year)
```
